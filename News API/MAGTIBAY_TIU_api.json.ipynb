```python
import pprint
import requests
import datetime
import pandas as pd

secret = "733dd542bffb474082e59bb4ea948919"
   
# Define the endpoint
url = 'https://newsapi.org/v2/everything?'
   
# Specify the query and
# number of returns
parameters = {
    'q': 'Philippines', # query phrase
    'pageSize': 10,  # maximum is 100
    'apiKey': secret, # your own API key
    'from': '2022-5-21',
    'to': '2022-5-22'
}
   
# Make the request
response = requests.get(url,  params = parameters)

response_json = response.json()
#pprint.pprint(response_json)

extracted_api_info = []

for i in response_json['articles']:
    title = i['title']
    author = i['author']
    publishedAt = i['publishedAt']
    fullArticle = i['content']
    extracted_api_info.append((title, author, publishedAt, fullArticle))
    
                              

```

```python
df_api_list = pd.DataFrame(extracted_api_info)
df_api_list.columns = ['Title', 'Author', 'PublishedAt', 'FullArticle']
df_api_list.to_json('MAGTIBAY_TIU_api.json',orient='index')
pd.read_json('MAGTIBAY_TIU_api.json')
```
